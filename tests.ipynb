## Simple test

```python
from MarkovMapMatching import MapGraph
from MarkovMapMatching import HMMMapMatcher
import matplotlib.pyplot as plt
import numpy as np


fig, ax = plt.subplots()

graph = MapGraph()
graph.addPoint("1", -99.18522930063155, 18.935347440600474)
graph.addPoint("2", -99.18508271809397, 18.936765014056512)
graph.addPoint("3", -99.18462022494504, 18.93671242301332)
graph.addPoint("4", -99.18476680746626,18.93523269550713)
graph.addPoint("5", -99.18414355237633,18.93505517643727)

graph.addConnection("1", "2")
graph.addConnection("2", "3")
graph.addConnection("3", "4")
graph.addConnection("1", "4")
graph.addConnection("4", "5")
graph.drawGraph(ax, nodeSize = 20)
```

```python
mapMatcher = HMMMapMatcher(graph, 4.07, 1)

point1 = [18.93604340952813, -99.18523258531134]
point1 = point1[::-1]
point2 = [18.937022395922813, -99.18481044326323]
point2 = point2[::-1]
point3 = [18.93590865944123, -99.18458655966684]
point3 = point3[::-1]


measurements = np.array([point1,point2, point3])
```

```python
path = mapMatcher.computePath(measurements)
print(path)
```

```python
cleanedMeasurements = mapMatcher.cleanMeasurements(measurements)
mapMatcher.drawMeasurementsAndPath(cleanedMeasurements, path, nodeSize = 3)
```

The green path is the one followed, and the blue dots are the points matched to the map.
